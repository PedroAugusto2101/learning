# Linear Regression: Cost Function

## üéØ Objectives
- Implement and understand the cost function for linear regression with one variable.
- Visualize the behavior of the cost function with different model parameters.

```python
# üìå Imports
import numpy as np
import matplotlib.pyplot as plt
```

## üóÇÔ∏è Dataset
A simple dataset of house sizes and their corresponding prices.

```python
x_train = np.array([1.0, 2.0])  # Size in 1000 sqft
y_train = np.array([300.0, 500.0])  # Price in $1000s
```

## üìâ Cost Function
Defined as:
$$ J(w, b) = \frac{1}{2m} \sum_{i=0}^{m-1} (wx^{(i)} + b - y^{(i)})^2 $$

```python
def compute_cost(x, y, w, b):
    m = x.shape[0]
    cost_sum = 0
    for i in range(m):
        prediction = w * x[i] + b
        cost = (prediction - y[i]) ** 2
        cost_sum += cost
    return cost_sum / (2 * m)
```

## üìä Cost Function Visualization
Plotting the cost for different values of `w` (keeping `b` fixed).

```python
w_values = np.linspace(100, 300, 100)
cost_values = [compute_cost(x_train, y_train, w, 100) for w in w_values]

plt.figure(figsize=(8, 4))
plt.plot(w_values, cost_values, label='Cost vs. w')
plt.xlabel('w')
plt.ylabel('Cost')
plt.title('Cost Function for Fixed b=100')
plt.legend()
plt.grid(True)
plt.show()
```

## üìà Expanded Dataset
More data points, imperfect linear fit.

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480, 430, 630, 730])

w_values = np.linspace(100, 300, 100)
cost_values = [compute_cost(x_train, y_train, w, 100) for w in w_values]

plt.figure(figsize=(8, 4))
plt.plot(w_values, cost_values)
plt.xlabel('w')
plt.ylabel('Cost')
plt.title('Cost for Larger Dataset (b = 100)')
plt.grid(True)
plt.show()
```

## ‚úÖ Conclusion
- The cost function measures the error of our model.
- Minimizing the cost helps us find the best parameters `w` and `b`.
