```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(data=cancer['data'], columns=cancer['feature_names'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
# we use StandardScaler to standardize the data
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
# after normalized data we will use the PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=cancer['target'], cmap='plasma')
plt.xlabel('Primeiro componente principal')
plt.ylabel('Segundo componente principal')
```

```python
pca.components_.shape
```

```python
pca.components_
# each value shows us the correlation between the original features and the principal components
# the first component is a linear combination of the original features
```

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
df_comp.head()
# when we do a PCA transformation we want to find out which combination of the original features is the most important for each component
# we can see that the first component is a linear combination of the original features
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_comp, cmap='plasma')
plt.xlabel('Features')
```
