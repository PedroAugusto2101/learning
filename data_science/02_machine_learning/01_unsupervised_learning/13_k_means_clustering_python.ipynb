```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
# make_blobs is a function to generate isotropic Gaussian blobs for clustering.
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=50)
# make_blobs(number_of_samples, number_of_features, number_of_centers, standard_deviation, random_state)
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap='rainbow')
# data[0][:, 0]: X-coordinates of the points (first feature).
# data[0][:, 1]: Y-coordinates of the points (second feature).
# c=data[1]: Colors the points based on their cluster labels.
# s=50: Sets the size of the points in the scatter plot.
# cmap='viridis': Specifies the colormap for coloring the points.
```

```python
from sklearn.cluster import KMeans
# KMeans is a clustering algorithm that partitions the data into K clusters.
```

```python
kmeans = KMeans(n_clusters=4)
# n_clusters=4: Specifies the number of clusters to form.
# cluster: a group of data points that are similar to each other.
```

```python
# as is a unsupervised algorithm, we don't need to split the data into train and test sets.
# we can fit the model directly to the data
kmeans.fit(data[0])
```

```python
# we also dont have the metrics to evaluate the model, so we can use the inertia metric.
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(12, 8))
# plot two images side by side
# sharey=True: Ensures that both plots share the same y-axis limits.
ax1.set_title('KMeans')
ax1.scatter(data[0][:, 0], data[0][:, 1], c=kmeans.labels_, cmap='rainbow')
# plot 1st image with the labels of the KMeans algorithm
ax2.set_title('Original')
ax2.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap='rainbow')
# plot 2nd image with the original labels of the data
```
